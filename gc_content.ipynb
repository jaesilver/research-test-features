```python
def gc_content(seq):
    gc_count = seq.count('G') + seq.count('C')
    total_count = len(seq)
    return (gc_count / total_count) * 100

def fasta_to_gc_content(fasta_file):
    with open(fasta_file) as f:
        seq = ''
        for line in f:
            if line.startswith('>'):
                if seq:
                    gc = gc_content(seq)
                    print(f'{gc:.2f}%\n')
                    seq = ''
                header = line.strip()
            else:
                seq += line.strip()
        if seq:
            gc = gc_content(seq)
#             print(f'Sequence: {header}\nGC content: {gc:.2f}%\n')
            print(f'{gc:.2f}%\n')
fasta_to_gc_content('lncRNA.fa')
```
